```julia
Revise.revise()

B = rand_tt(1, 4, 2)
@show normalization(B)
# B.z = exp(-100)
# absorb_z_into_matrices!(B)
# B = B + identity_tensor_train(B)
B_inv = GlauberPT.inverse_tt_old(B, 3; steps=1000)
# (B_inv, z_acc) = inverse_tt(B, 30; steps=20)
# B_inv.z *= z_acc

X = [[rand(1:2) for i in 1:length(B)] for _ in 1:1000]
error_sample = maximum(abs.([evaluate(B,x) for x in X] .- [1/evaluate(B_inv,x) for x in X]))
avg_error = mean(abs.([evaluate(B,x) for x in X] .- [1/evaluate(B_inv,x) for x in X]))
println(avg_error)
println(error_sample)
Bx = [evaluate(B,x) for x in X]
B_invx = [evaluate(B_inv,x) for x in X]
@assert !any(any(isinf, bx) for bx in Bx)
@assert !any(any(isinf, b_invx) for b_invx in B_invx)
@assert !any(any(isnan, bx) for bx in Bx)
@assert !any(any(isnan, b_invx) for b_invx in B_invx)
# scatter(Bx .* B_invx)
xmin = argmin(Bx)
xmax = argmax(Bx)
linex = [Bx[xmin], Bx[xmax]]
liney = [1/B_invx[xmin], 1/B_invx[xmax]]
plot(linex, liney, color=:black)
scatter!(Bx, 1 ./ B_invx)
# Bx .* B_invx
```
